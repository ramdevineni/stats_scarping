```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

import numpy as np

url='https://www.bikedekho.com/new-bikes'
page=urlopen(url)

page_html=page.read()
page.close()
soup=BeautifulSoup(page_html,'html.parser')

print(soup.head.title.text)

bikes_urls=soup.findAll("div",{"id":"all_brands"})
print(len(bikes_urls))
bikes_url=bikes_urls[0].findAll('ul',{'class':'listing gsc_row clearfix'})
bikes_url=bikes_url[0].findAll('li',{'class':'gsc_col-xs-4 gsc_col-sm-3 gsc_col-md-3 gsc_col-lg-2'})
specs_dict={'brand':[],'Mileage':[],'Displacement':[],'No. of Cylinders':[],'Max Power':[],'Max Torque':[],'Front Brake':[],'Rear Brake':[],'rating':[],'reviews':[],'bike_names':[]}


print(len(bikes_url))
for x in bikes_url:
    y= x.findAll('a')
    z=y[0].get('href')
    url_for_brand='https://www.bikedekho.com'+z
    page_brand=urlopen(url_for_brand)
    brand_html=page_brand.read()
    page_brand.close()
    soup_brand=BeautifulSoup(brand_html,'html.parser')
    #print(soup_brand.head.title.text)
    bikes_brand=soup_brand.findAll('li',{'class':'gsc_col-xs-12 gsc_col-sm-6 gsc_col-md-12 gsc_col-lg-12'})

    bikes_info_list=[[],[],[],[],[],[],[],[],[],[]]
    list1=[]
    list2=[]
    for b in bikes_brand:
        name=b.findAll('a')
        #print('href length '+ str(len(name)))
        #print(name[0].get('href'))

        try:
            bike_specs_url = 'https://www.bikedekho.com'+name[0].get('href')+'/specifications'
            print(bike_specs_url)

            specs_page=urlopen(bike_specs_url)
            specs_html=specs_page.read()
            specs_page.close()
            specs_soup=BeautifulSoup(specs_html,'html.parser')

            #print(len(specs_soup.findAll('div',{'class':'title'})))

            specs_list=['Displacement','No. of Cylinders','Max Power','Max Torque','Front Brake','Rear Brake']


            review=specs_soup.findAll('span',{'class':'ratingvalue'})
            #print(float(review[0].text.split('/')[0]))
            try:
                specs_dict['rating'].append(float(review[0].text.split('/')[0]))
            except:
                specs_dict['rating'].append(np.NaN)
            reviews_count=specs_soup.findAll('span',{'class':'bottomText'})
            try:
                specs_dict['reviews'].append(int(reviews_count[0].text.split(' ')[0]))
            #print(int(reviews_count[0].text.split(' ')[0]))
            except:
                specs_dict['reviews'].append(np.NaN)
            specs_dict['brand'].append(y[0].findAll('span')[0].text)
            print(y[0].findAll('span')[0].text)


            print(specs_soup.head.title.text)
            try:
                specs_dict['bike_names'].append(specs_soup.head.title.text.split('Specifications')[0].rstrip(' '))
            except:
                specs_dict['bike_names'].append(specs_soup.head.title.text)
                #print('yes')
            key_specs=specs_soup.findAll('table',{'class':'keyfeature'})
            #print(len(key_specs))
            key_spec1=key_specs[0]
            rows=key_spec1.findAll('tr')
            row1=rows[0]
            if re.findall('Mileage|mileage',row1.findAll('td')[0].text):
                try:
                    #print(float(row1.findAll('td')[1].text.split(' ')[0]))
                    specs_dict['Mileage'].append(float(row1.findAll('td')[1].text.split(' ')[0]))
                except:
                    specs_dict['Mileage'].append(np.NaN)
            l=len(rows)
            for r in range(1,l):
                #print(len(row.findAll('td')))
                # for td in row.findAll('td'):
                #     print(td.text)

                if specs_list.count(rows[r].findAll('td')[0].text)!=0:
                    specs_list.remove(rows[r].findAll('td')[0].text)
                    #print(rows[r].findAll('td')[1].text)


                    try:
                        specs_dict[rows[r].findAll('td')[0].text].append(float(rows[r].findAll('td')[1].text.replace(' ','').split(' ')[0].split('c')[0]))
                    except:
                        specs_dict[rows[r].findAll('td')[0].text].append(rows[r].findAll('td')[1].text.lstrip(' ').rstrip(' '))
            for i in range(len(specs_list)):
                specs_dict[specs_list[i]].append(np.NaN)





        except:
            pass

    for i in specs_dict:
        print(i)
        print(len(specs_dict[i]),specs_dict[i].count(np.NaN))
```

```python
import pandas as pd
df=pd.DataFrame(columns=['brand','bike_names','Mileage','Displacement','No. of Cylinders','Max Power','Max Torque','Front Brake','Rear Brake','rating','reviews'])
for column in df.columns:
    df[column]=specs_dict[column]
df
```

```python
df.to_csv('bikes_data.csv',index=False)
```
